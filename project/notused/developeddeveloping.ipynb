```python
stocks=pd.read_csv('developingdeveloped.csv')
stocks.head(2)
```

```python
country = stocks['Country'].values
commodity= stocks['Commodity'].values
year=stocks['TIME'].values
value=stocks['Value'].values
variable=stocks['Variable'].values
```

```python
countries=['Australia','Canada','Chile','Colombia','United Kingdom','Israel','Japan','Korea','Mexico','Norway','New Zealand','Switzerland','United States','Russian Federation','Egypt','Ethiopia','Nigeria','Argentina','Brazil','Paraguay','Peru','China','India','Indonesia','Iran','Kazakhstan']
```

```python
df = pd.DataFrame({"country":country, "year":year, "foodvalue":value,"meat":commodity})
df = df[df.country == 'DEVELOPING COUNTRIES']
df = df[df['meat'].isin(['Beef and veal','Poultry meat','Sheepmeat','Pigmeat'])]
```

```python
df1 = pd.DataFrame({"country":country, "year":year, "foodvalue":value,"meat":commodity})
df1 = df1[df1.country == 'DEVELOPED COUNTRIES']
df1 = df1[df1['meat'].isin(['Beef and veal','Poultry meat','Sheepmeat','Pigmeat'])]
```
